# Exploration du dataset "Superstore Sales"

## 1. Chargement

```python
import pandas as pd

df = pd.read_csv("data/raw/superstore_sales.csv")

print(df.info())
print()
print(df.head())
```

## 2. Standardisation des colonnes et dates

```python
df.columns = df.columns.str.lower().str.strip()
df['order date'] = pd.to_datetime(df['order date'], dayfirst=True)
df['ship date']  = pd.to_datetime(df['ship date'],  dayfirst=True)
```

## 3. Valeurs manquantes

```python
print("\nValeurs manquantes :")
print(df.isnull().sum())
df['postal code'] = df['postal code'].fillna(0).astype(int)
```

## 4. Doublons

```python
print(f"\nDoublons : {df.duplicated().sum()}")
print(f"Commandes uniques (Order ID) : {df['order id'].nunique()}")
```

## 5. Valeurs aberrantes sur Sales

```python
print("\nStatistiques Sales :")
print(df['sales'].describe())
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[df['sales'] > Q3 + 1.5 * IQR]
print(f"Valeurs aberrantes détectées : {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
```

## 6. Distribution des variables clés

```python
print("\nRépartition par catégorie :")
print(df['category'].value_counts())
print("\nRépartition par région :")
print(df['region'].value_counts())
```

## 7. Export

```python
import os
os.makedirs("data/out", exist_ok=True)
df.to_csv("data/out/ventes_clean.csv", index=False)
print("Fichier ventes_clean.csv exporté")
```
