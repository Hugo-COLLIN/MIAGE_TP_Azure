```python
import pandas as pd

df = pd.read_csv("data/superstore_sales.csv")

# 1. Visualiser les premières lignes
print(df.head())
print(df.info())

# 2. Identifier les valeurs manquantes
print(df.isnull().sum())

# 3. Identifier les outliers (ex: montant des ventes)
print(df.describe())

# 4. Standardiser les valeurs (ex: minuscules, format date)
df.columns = df.columns.str.lower().str.strip()
df['date'] = pd.to_datetime(df['date'])
df.dropna(inplace=True)

df.to_csv("ventes_clean.csv", index=False)
```
